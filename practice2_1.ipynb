```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt

from google.colab import drive

drive.mount('/content/gdrive')
# 
origin_img = cv2.imread('/content/gdrive/My Drive/Image_Processing/lena.jpg')

# BGR RGB 
RGB_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)
# R,G,B  
Red_img, Green_img, Blue_img = cv2.split(RGB_img)
#  
plt.figure( figsize=(20,20) ) #   
plt.subplot(1, 4, 1) # 1 4 1 
#  
plt.title("Original Image")
plt.imshow(RGB_img)
# R 
RGB_img[:,:,0] = Red_img
RGB_img[:,:,1] = 0
RGB_img[:,:,2] = 0
plt.axis("off")
plt.subplot(1, 4, 2) # 1 4 2 
plt.title("Red Channel")
plt.imshow(RGB_img)
# G  
RGB_img[:,:,0] = 0
RGB_img[:,:,1] = Green_img
RGB_img[:,:,2] = 0
plt.axis("off")
plt.subplot(1, 4, 3) # 1 4 3 
plt.title("Green Channel")
plt.imshow(RGB_img)
# B  
RGB_img[:,:,0] = 0
RGB_img[:,:,1] = 0
RGB_img[:,:,2] = Blue_img
plt.axis("off")
plt.subplot(1, 4, 4) # 1 4 4 
plt.title("Blue Channel")
plt.imshow(RGB_img)
plt.axis("off")
plt.show()
```
