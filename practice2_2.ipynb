```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt

from google.colab import drive

drive.mount('/content/gdrive')
# 
origin_img = cv2.imread('/content/gdrive/My Drive/Image_Processing/lena.jpg')
# BGR RGB 
RGB_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)
# RGB HSV 
HSV_img = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2HSV)

# H, S, V  
H_img, S_img, V_img = cv2.split(HSV_img)

#  
plt.figure( figsize=(20,20) ) #   
plt.subplot(1, 4, 1) # 1 4 1  plt.title("First image")
plt.title("Original Image")
plt.imshow(RGB_img)
plt.axis("off")
plt.subplot(1, 4, 2) # 1 4 2 
RGB_img[:,:,0] = H_img
RGB_img[:,:,1] = H_img
RGB_img[:,:,2] = H_img
plt.title("H Channel")
plt.imshow(RGB_img)
plt.axis("off")
plt.subplot(1, 4, 3) # 1 4 3 
RGB_img[:,:,0] = S_img
RGB_img[:,:,1] = S_img
RGB_img[:,:,2] = S_img
plt.title("S Channel")
plt.imshow(RGB_img)
plt.axis("off")
plt.subplot(1, 4, 4) # 1 4 4 
RGB_img[:,:,0] = V_img
RGB_img[:,:,1] = V_img
RGB_img[:,:,2] = V_img
plt.title("V Channel")
plt.imshow(RGB_img)
plt.axis("off")
plt.show()
```
